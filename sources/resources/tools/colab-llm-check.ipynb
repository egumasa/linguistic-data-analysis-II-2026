# Colab LLM check

*Linguistic Data Analysis II — a 1-minute check that the LLM works on **your** account.*

Colab has a built-in Gemini model you can call with `from google.colab import ai`. Before Day 1 we need to confirm it works when **you** are signed in with your Tohoku Google account.

**What to do:** **Runtime → Run all**, wait for every cell to finish, then copy the final ✅ / ❌ message and report it to the instructor.

> This notebook only runs **inside Google Colab** — `google.colab.ai` does not exist on a normal computer.

```python
# Step 1 — load Colab's built-in Gemini (no API key needed)
from google.colab import ai

def run(prompt):
    """Send one prompt to the model and return its reply as text."""
    return ai.generate_text(prompt)

print("Loaded google.colab.ai — ready to try a prompt.")
```

```python
# Step 2 — try a simple prompt and read the reply
reply = run("Say hello in one short sentence.")
print(reply)
```

## Have some fun 🎉

The model can do more than say hello. Change the **topic** in the form on the right (in Colab), then run the cell and see what it comes up with. Run it a few times — the answer changes each time!

```python
#@title 🎈 Ask the model for a haiku { display-mode: "form" }
topic = "cats learning statistics"  #@param {type:"string"}

haiku = run(f"Write a short, funny haiku about: {topic}")
print(haiku)

# Try changing `topic` above to anything you like — your hometown, your favorite
# food, corpus linguistics... then run this cell again.
```

```python
# One more for fun: ask it to explain something as if you were five.
print(run("Explain what a 'corpus' is in linguistics, in one sentence a 5-year-old would get."))
```

```python
# Step 3 — final check: did the LLM run?  Copy the message below and report it.
try:
    answer = run("What is 2 + 2? Reply with just the number.")
    print("✅ SUCCESS — Colab AI works on my account. (model replied:", repr(answer.strip()), ")")
except Exception as e:
    print("❌ FAILED — Colab AI did not run on my account.")
    print("   error:", type(e).__name__, "-", e)
    print("   → Copy this whole message and send it to the instructor.")
```
