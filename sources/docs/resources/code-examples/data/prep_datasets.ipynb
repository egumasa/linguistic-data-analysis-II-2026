## Validate all gold files against the canonical schema


```python
import json, glob, os
for p in sorted(glob.glob("gold/*.json")):
    d = json.load(open(p))
    ids = [x["id"] for x in d]
    assert len(ids) == len(set(ids)), f"{p}: duplicate ids"
    assert all(x["text"].strip() and x["label"] for x in d), f"{p}: empty field"
    labels = sorted(set(x["label"] for x in d))
    print(f"{os.path.basename(p):28} {len(d):5} items  {labels}")
print("ALL VALID")
```
